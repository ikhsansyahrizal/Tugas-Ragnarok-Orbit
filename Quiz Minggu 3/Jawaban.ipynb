```python
#contoh implenetasi OOP

from abc import abstractmethod, ABC

#Abstaction
class Pakaian(ABC):
    def tagline(self, merek):
        print("Local Pride: ", merek)



#Encapsulation & Inheritance
class Baju(Pakaian):
    def __init__(self, imerek, iukuran, iwarna):
        self.imerek = imerek
        self.iukuran = iukuran
        self.iwarna = iwarna

    
    def info_pakaian(self):
        print("""
        Merek Pakaian : %s
        Ukuran Pakaian : %s
        Warna Pakaian : %s
        """ %(self.imerek, self.iukuran, self.iwarna))


baju_erigo1 = Baju("Erigos","L","Krem")
baju_erigo1.tagline("Erigo")
baju_erigo1.info_pakaian()



#nomor 3
import math
x = math.sqrt(175) + 4*1 / math.sqrt(7) + 1/3 + math.log(81, 3)
print(x)



#nomor 4
class NegativeError(Exception):
   def __init__(self, data):
       self.data = data
try:
   x = -2
   if x < 0:
       raise NegativeError(x)
except NegativeError as e:
   print("Nilai {}. adalah negatif, masukkan nilai positif".format(e))


```

<h1>Machine Learning</h1>
1. Beda Supervised dan Unsupervisied Learning
- Supervised
    Machine Learning model yang mempelajari data dengan label atau target dimana evaluasi model tersebut akan berdasarkan target ini.
    contohnya : identifikasi wajah, klasifikasi gambar

<br><br>
- Unsupervisied
    Machine Learning Model yang mempelajari pola data tanpa adanya target data
    contoh : clustering data pelanggan, deteksi anomali

<br><br>
2. ML Lifecylce: 
-   1. Tentukan Tujuan -> tentukan tujuan yang akan dicapai
    2. Kumpulkan Data -> kumpulkan data yang diperlukan dalam pembuatan model ML
    3. Siapkan Data -> data yang ada disiapkan untuk dimasukkan ke model
    4. Pilih Algoritma -> memilih algoritma yang akan di implementasikan pada model sesuai dengan kasus dan tujuan.
    5. Latih Model -> melatih model menggunakan data yang sudah tersedia.
    6. Tes Model -> model yang sudah dilatih masih perlu di tes apakah output yang dihasilkan sudah sesuai tujuan.
    7. Integrasikan Model -> model yang sudah siap, di integrasikan pada teknologi. Mobile / Web / dll.

<br><br>
3. Beda Regression dan Clustering
- Regression
    teknik untuk identifikasi relasi / hubungan diantara 2 variabel atau lebih.

- Clustering
    teknik untuk mengelompokkan suati dataset ke dalam beberapa kelompok berdasarkan kemiripan pada objek data dan sebaliknya meminimalkan kemiripan terhadap kluster yang lain.
