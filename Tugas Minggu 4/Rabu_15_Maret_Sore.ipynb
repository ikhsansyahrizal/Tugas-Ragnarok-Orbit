```python
#divide data
exited = bank_churn.loc[bank_churn['Exited'] == 1 ]
not_exited = bank_churn.loc[bank_churn['Exited'] == 0 ]
```

<hr/>

<h1> Bivariate Data Analysis </h1>


```python
#setup matplotlib figure
f, axes = plt.subplots(ncols = 1, figsize=(5,5), sharex=True)
sns.despine(left=True)

sns.histplot(data=exited["Age"],label="exited",color='b',kde=True,linewidth=2,alpha=0.3).set_title('Age vs Exited')
sns.histplot(data=not_exited["Age"],label="not_exited",color='r',kde=True,linewidth=2,alpha=0.3)
```

```python

sns.histplot(data=exited["EstimatedSalary"],label="exited",color='b',kde=True,linewidth=2,alpha=0.3).set_title('Salary vs Exited')
sns.histplot(data=not_exited["EstimatedSalary"],label="not_exited",color='r',kde=True,linewidth=2,alpha=0.3)
```

```python

sns.histplot(data=exited["Tenure"],label="exited",color='b',kde=True,linewidth=2,alpha=0.3).set_title(' Tenure vs Exited')
sns.histplot(data=not_exited["Tenure"],label="not_exited",color='r',kde=True,linewidth=2,alpha=0.3)
```

```python

sns.histplot(data=exited["Balance"],label="exited",color='b',kde=True,linewidth=2,alpha=0.3).set_title('Balance vs Exited')
sns.histplot(data=not_exited["Balance"],label="not_exited",color='r',kde=True,linewidth=2,alpha=0.3)
```

```python
sns.pairplot(bank_churn, hue="Exited", diag_kind="hist")
plt.show()
```
